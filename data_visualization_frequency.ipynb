```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
usa_url = 'https://raw.githubusercontent.com/caroline710/big-data-final/main/data/distilbert-usa.csv?token=GHSAT0AAAAAACKS7T4CGFH23ZXDXQEL7ARMZLSJJCA'
usa_data = pd.read_csv(usa_url)
usa_data.head()
```

```python
usa_data['Year'] = pd.DatetimeIndex(usa_data['Month_Year']).year
usa_data['Month'] = pd.DatetimeIndex(usa_data['Month_Year']).month_name()
usa_data.head()
```

```python
usa_year_groups = usa_data.groupby('Year')
usa_data_2018 = usa_year_groups.get_group(2018)
usa_data_2019 = usa_year_groups.get_group(2019)
usa_data_2020 = usa_year_groups.get_group(2020)
usa_data_2020.head()
```

```python
eu_url = 'https://raw.githubusercontent.com/caroline710/big-data-final/main/data/distilbert-eu.csv?token=GHSAT0AAAAAACKS7T4DDFKMSWZRVAAQHYLGZLSJJTA'

eu_data = pd.read_csv(eu_url)
eu_data.head()
```

```python
eu_data['Year'] = pd.DatetimeIndex(eu_data['Month_Year']).year
eu_data['Month'] = pd.DatetimeIndex(eu_data['Month_Year']).month_name()
eu_data.head()
```

```python
eu_year_groups = eu_data.groupby('Year')
eu_data_2018 = eu_year_groups.get_group(2018)
eu_data_2019 = eu_year_groups.get_group(2019)
eu_data_2020 = eu_year_groups.get_group(2020)
eu_data_2020.head()
```

```python
sns.set_theme(style='whitegrid')
# 2018 data
plt.figure(figsize=(12.5,6))
sns.lineplot(data=usa_data_2018, x='Month', y='SentimentScore', color='plum', marker='o', label='usa')
sns.lineplot(data=eu_data_2018, x='Month', y='SentimentScore', color='mediumseagreen', marker='o', label='eu')
plt.legend(loc='upper right', fontsize='large')
plt.title('2018', size=20)
plt.show()
```

```python
# 2019 data
plt.figure(figsize=(12.5,6))
sns.lineplot(data=usa_data_2019, x='Month', y='SentimentScore', color='plum', marker='o', label='usa')
sns.lineplot(data=eu_data_2019, x='Month', y='SentimentScore', color='mediumseagreen', marker='o', label='eu')
plt.legend(loc='upper right', fontsize='large')
plt.title('2019', size=20)
plt.show()
```

```python
# 2020 data
plt.figure(figsize=(12.5,6))
sns.lineplot(data=usa_data_2020, x='Month', y='SentimentScore', color='plum', marker='o', label='usa')
sns.lineplot(data=eu_data_2020, x='Month', y='SentimentScore', color='mediumseagreen', marker='o', label='eu')
plt.legend(loc='upper right', fontsize='large')
plt.title('2020', size=20)
plt.show()
```

```python
# usa hist plot
plt.figure(figsize=(12.5,6))
sns.histplot(data=usa_data, x='SentimentScore', color='plum')
plt.title('USA (2018-2020)')
```

```python
# eu hist plot
plt.figure(figsize=(12.5,6))
sns.histplot(data=eu_data, x='SentimentScore', color='mediumseagreen')
plt.title('EU (2018-2020)')
```
